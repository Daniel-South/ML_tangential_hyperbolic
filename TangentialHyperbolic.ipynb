```python
# Gain intuition by demonstrating the output of the tangential hyperbolic function over a range of input values.

# tanh(x) = (e^x - e^-x) / (e^x + e^-x)

# The output of tanh(x) is centered around zero and approaches -1 and 1 asymptotically.

# tanh(x) is used as an activation fuction in Recurrent Neural Networks, i.e. neural networks designed
# to process sequence data, such as stock prices or sequences of words, where order is significant.

# The zero-centered output of tanh(x) helps networks converge and learn faster.

# Daniel R. South
# 2026-06-13
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("Generate a list of x values from -5 to 5")
x1 = np.linspace(-5, 5, 21)
print(x1)

y1 = np.tanh(x1)
print(y1)


print("\nGenerate a list of x values from -2 to 2")
x2 = np.linspace(-2, 2, 21)
print(x2)

y2 = np.tanh(x2)
print(y2)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(6,10))
ax1.plot(x1, y1)
ax1.set_title("tanh function, -5 <= x <= 5")
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2.plot(x2, y2)
ax2.set_title("tanh function, -2 <= x <= 2")
ax2.set_xlabel("x")
ax2.set_ylabel("y")

plt.tight_layout()
plt.show()
```
